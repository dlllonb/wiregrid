```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = "../data"

print("Files in ../data:")
for f in os.listdir(data_dir):
    print(f)
```

```python
filename = "beam_map_331F1_pol1.grd"
filepath = os.path.join("../data", filename)

with open(filepath, "r") as f:
    lines = f.readlines()

for i, line in enumerate(lines):
    if line.strip():
        parts = line.split()
        if len(parts) == 3:
            try:
                nx, ny, nz = map(int, parts)
                if nz == 0:
                    grid_line = i
                    break
            except:
                pass

print("Grid line:", grid_line)
print("Grid size:", nx, ny, nz)

data_lines = lines[grid_line + 1:]
rows = []

for line in data_lines:
    parts = line.split()
    if len(parts) == 4:
        rows.append([float(x) for x in parts])

arr = np.array(rows)
print("Data shape:", arr.shape)
```

```python
img = arr[:, 0].reshape(ny, nx)

plt.imshow(img, origin="lower")
plt.colorbar()
plt.title(filename + " (column 1)")
plt.show()
```
